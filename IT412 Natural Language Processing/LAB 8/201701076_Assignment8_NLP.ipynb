```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
from sumy.summarizers.lex_rank import LexRankSummarizer
from sumy.summarizers.text_rank import TextRankSummarizer
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words
```

```python
path = 'C:/Users/Vedant/Desktop/NLP LAB/Assignment8/Assignment-8 Dataset/'
```

```python
import os
import io
import pandas as pd
import numpy as np
```

```python
%cd C:/Users/Vedant/Desktop/NLP LAB/Assignment8/Assignment-8 Dataset/News article/
```

```python
files = os.listdir()
stemmer = Stemmer('english')
summarizer = LexRankSummarizer(stemmer)
summarizer1 = TextRankSummarizer(stemmer)
for i in range(len(files)):
    parser = PlaintextParser.from_file(os.path.join(path,'News article/',files[i]),Tokenizer('english'))
    
    s = ""
    for sentence in summarizer(parser.document, 2):
        s = s + str(sentence)
    with open(os.path.join(path,'results_2_l/',files[i]),'a') as f:
        f.write(s)
    s = ""
    for sentence in summarizer1(parser.document, 2):
        s = s + str(sentence)
    with open(os.path.join(path,'results_2/',files[i]),'a') as f:
        f.write(s)
    
        s = ""
    for sentence in summarizer(parser.document, 3):
        s = s + str(sentence)
    with open(os.path.join(path,'results_3_l/',files[i]),'a') as f:
        f.write(s)
    s = ""
    for sentence in summarizer1(parser.document, 3):
        s = s + str(sentence)
    with open(os.path.join(path,'results_3/',files[i]),'a') as f:
        f.write(s)
        
        s = ""
    for sentence in summarizer(parser.document, 4):
        s = s + str(sentence)
    with open(os.path.join(path,'results_4_l/',files[i]),'a') as f:
        f.write(s)
    s = ""
    for sentence in summarizer1(parser.document, 4):
        s = s + str(sentence)
    with open(os.path.join(path,'results_4/',files[i]),'a') as f:
        f.write(s)
```

```python
path_r = 'C:/Users/Vedant/Desktop/NLP LAB/Assignment8/Assignment-8 Dataset/Summary'
```

```python
from sumy.evaluation.rouge import rouge_1
from sumy.evaluation.rouge import rouge_2
from sumy.models.dom._sentence import Sentence
```

```python
r1l_2 = []
r2l_2 = []
r1l_3 = []
r2l_3 = []
r1l_4 = []
r2l_4 = []
r1t_2 = []
r2t_2 = []
r1t_3 = []
r2t_3 = []
r1t_4 = []
r2t_4 = []
```

```python
for i in range(len(files)):
    with open(os.path.join(path,'results_2',files[i]),'r+') as f:
        s = f.read()
        s = s.split('.')
        s = [Sentence(x,Tokenizer('english')) for x in s]
    with open(os.path.join(path,'results_2_l',files[i]),'r+') as f:
        s1 = f.read()
        s1 = s1.split('.')
        s1 = [Sentence(x,Tokenizer('english')) for x in s1]
    with open(os.path.join(path,'results_3',files[i]),'r+') as f:
        s2 = f.read()
        s2 = s2.split('.')
        s2 = [Sentence(x,Tokenizer('english')) for x in s2]
    with open(os.path.join(path,'results_3_l',files[i]),'r+') as f:
        s3 = f.read()
        s3 = s3.split('.')
        s3 = [Sentence(x,Tokenizer('english')) for x in s3]
    with open(os.path.join(path,'results_4',files[i]),'r+') as f:
        s4 = f.read()
        s4 = s4.split('.')
        s4 = [Sentence(x,Tokenizer('english')) for x in s4]
    with open(os.path.join(path,'results_4_l',files[i]),'r+') as f:
        s5 = f.read()
        s5 = s5.split('.')
        s5 = [Sentence(x,Tokenizer('english')) for x in s5]
    with open(os.path.join(path_r,files[i]),'r+') as f:
        sr = f.read()
        sr = sr.split('.')
        sr = [Sentence(x,Tokenizer('english')) for x in sr]
        
    r1l_2.append(rouge_1(s1,sr))
    r1l_3.append(rouge_1(s3,sr))
    r1l_4.append(rouge_1(s5,sr))
    r2l_2.append(rouge_2(s1,sr))
    r2l_3.append(rouge_2(s3,sr))
    r2l_4.append(rouge_2(s5,sr))
    r1t_2.append(rouge_1(s,sr))
    r1t_3.append(rouge_1(s2,sr))
    r1t_4.append(rouge_1(s4,sr))
    r2t_2.append(rouge_2(s,sr))
    r2t_3.append(rouge_2(s2,sr))
    r2t_4.append(rouge_2(s4,sr))    
```

```python
d_Lex = {'file':files,'rouge_1_Lex_2':r1l_2,'rouge_2_Lex_2':r2l_2,'rouge_1_Lex_3':r1l_3,'rouge_2_Lex_3':r2l_3,'rouge_1_Lex_4':r1l_4,'rouge_2_Lex_4':r2l_4}
d_text = {'file':files,'rouge_1_text_2':r1t_2,'rouge_2_text_2':r2t_2,'rouge_1_text_3':r1t_3,'rouge_2_text_3':r2t_3,'rouge_1_text_4':r1t_4,'rouge_2_text_4':r2t_4}
```

```python
df_Lex = pd.DataFrame(data = d_Lex)
df_text = pd.DataFrame(data = d_text)
```
