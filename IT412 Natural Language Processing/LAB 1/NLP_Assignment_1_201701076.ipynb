```python
from bs4 import BeautifulSoup
import re
import nltk
import pandas as pd
```

```python
with open('wsj8_001') as f:
   soup = BeautifulSoup(f, "html.parser")
```

```python
Texts = {}
i = 0
for tag in soup.find_all("text"):
    Texts[str(i)] = tag.text
    i = i+1
```

```python
D = Texts['0']
```

```python
print(D)
```

Remove Unwanted Whitespaces

```python
def remove_whitespaces(text):
    x = text.split()
    return ''.join(w+" " for w in x)
```

```python
D1 = remove_whitespaces(D)
```

Remove Contractions

```python
def decontracted(phrase):
    # specific
    phrase = re.sub(r"won\'t", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)
    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```python
D2 = decontracted(D1)
print(D2)
```

Removing punctuation from titles and contents

```python
from string import punctuation
def strip_punctuation(s):
    return ''.join(c for c in s if c not in punctuation)
```

```python
D3 = strip_punctuation(D2)
print(D3)
```

Remove Numbers

```python
def remove_nos(text):
     return ''.join(c for c in text if not c.isdigit())
```

```python
D4 = remove_nos(D3)
```

Tokenization

```python
D5 = nltk.word_tokenize(D4)
for i in range(len(D5)):
    D5[i] = D5[i].lower()
```

Stopwords

```python
from nltk.corpus import stopwords
stoplist = set(stopwords.words("english"))
for w in D5:
    if w not in stoplist:
        D6.append(w)
```

Stemming

```python
from nltk.stem import PorterStemmer 
ps = PorterStemmer()
D7 = ""
for w in D6:
    D7 = D7 + ps.stem(w) + " "
print(D7)
```

Lemmentation

```python
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer()
D8 = ""
for w in D6:
    D8 = D8 + ps.stem(w) + " "
print(D8)
```
